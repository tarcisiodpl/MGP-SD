### Gerando um arquivo .txt vazio por autor

```python
#funo para remover acentos
from unicodedata import normalize
def remover_acentos(txt):
    return normalize('NFKD', txt).encode('ASCII', 'ignore').decode('ASCII')

autoresList = []
for index, row in titulos_com_autores.iterrows():
    titulo, com, ID, autores = row
    autores = autores.replace(', ', ',')
    autoresList = autores.split(",")
    #print(autoresList)
    for a in autoresList:
        a = remover_acentos(a) #removendo acendo: ex.  -> a
        a = re.sub('[^a-zA-Z]', '', a) #mantendo letras e nmeros: ex. Jose S. -> JoseS
        print('Salvando arquivo: ' + a)
        nome = 'txt_autores\\' + a + '.txt'
        file = open(nome,"w+")
        #file.write(texto)
        file.close()  
```

### Salvando contedos dos artigos no arquivo de cada autor

```python
autoresList = []
for index, row in titulos_com_autores.iterrows():
    titulo, com, ID, autores = row
    autores = autores.replace(', ', ',')
    autoresList = autores.split(",")
    
    #capturando contedo do artigo
    artigo = 'txt_semRefTitulo\\' + ID + '.txt' 
    print('abrindo arquivo ' + artigo)
    file = open(artigo,"r",encoding='utf-8')
    texto = file.read()
    #print(texto)
    file.close()  
    #salvando contedo do artigo no arquivo de cada autor
    for a in autoresList:  
        a = remover_acentos(a) #removendo acendo: ex.  -> a
        a = re.sub('[^a-zA-Z]', '', a) #mantendo letras e nmeros: ex. Jose S. -> JoseS
        nome = 'txt_autores\\' + a + '.txt'
        print('salvando em ' + nome)
        file = open(nome,"a",encoding='utf-8')
        file.write(texto)
        file.write('\n')
        file.close()  
```
