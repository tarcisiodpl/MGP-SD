# Formando base com todos os autores

### Base com todos os autores e as respectivas palavras utilizadas por eles

```python
doc_ter_matrix = pd.read_csv('C:\\Users\\CoffeeLake_01\\Documents\\GitHub\\MGP-SD\\bracis\\doc_term_matrix_tfidf_with_header_stem_discrete.csv',
                  sep = ';')
doc_ter_matrix.head(10) 
```

### Base contendo os autores das 10 comunidades mais relevantes e as comunidades as quais eles percentem

```python
nosRede = pd.read_csv('C:\\Users\\CoffeeLake_01\\Documents\\GitHub\\MGP-SD\\bracis\\NosRede_c_ComunidadesFILTRO.csv',
                  sep = ',')
nosRede.head(10)
```

```python
nosRede.shape
```

### Merge formando base de dados conde cada linha representa as palavras utilizadas pelos autores das 10 comunidades mais relevantes e o rÃ³tulo Ã© a comunidade do autor

```python
authorWord_community = pd.merge(doc_ter_matrix,nosRede, left_on='Unnamed: 0', right_on='Id') 
authorWord_community = authorWord_community.drop(columns=['Unnamed: 0', 'Id', 'Label', '0'])
authorWord_community.head(10)
```

```python
authorWord_community.shape
```

### Salvando base em arquivo .csv

```python
authorWord_community.to_csv('bracis_resumoTitulo.csv', sep=',', index=False)
```

```python
authorWord_community.modularity_class.unique()
```
